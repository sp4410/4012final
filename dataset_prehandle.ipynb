```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('NYC_Weather_2016_2022.csv')

# Ensure the 'time' column is in datetime format
df['time'] = pd.to_datetime(df['time'], errors='coerce')

# Filter for data from July or August of 2022
df_2022_summer = df[
    (df['time'].dt.year == 2022) &
    (df['time'].dt.month.isin([7, 8]))
]
```

```python
df_2022_summer.to_csv("weather.csv",index=False)
```
